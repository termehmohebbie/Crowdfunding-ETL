Create the Category and Subcategory DataFrames
Create a Category DataFrame that has the following columns:

A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
A "category" column that has only the categories.
Export the DataFrame as a category.csv CSV file.

Create a SubCategory DataFrame that has the following columns:

A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
A "subcategory" column that has only the subcategories.
Export the DataFrame as a subcategory.csv CSV file.

```python
# Get the crowdfunding_info_df columns.
crowdfunding_info_df.columns
```

```python

crowdfunding_info_df.head()
```

```python
# Assign the category and subcategory values to category and subcategory columns.
crowdfunding_info_df['category'] = [crowdfunding_info_df['category & sub-category'][i].split('/')[0] for i in range(len(crowdfunding_info_df['category & sub-category']))]
crowdfunding_info_df['subcategory'] = [crowdfunding_info_df['category & sub-category'][i].split('/')[1] for i in range(len(crowdfunding_info_df['category'])) ]
crowdfunding_info_df.head()
```

```python
# Get the unique categories and subcategories in separate lists.
categories = [category for category in crowdfunding_info_df['category'].unique()]
subcategories = [subcat for subcat in crowdfunding_info_df['subcategory'].unique()]
print(categories)
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id. 

# Use a list comprehension to add "subcat" to each subcategory_id.    
cat_ids = [f'cat{cat_id}' for cat_id in category_ids]
scat_ids = [f'subcat{scat_id}' for scat_id in subcategory_ids] 
print(cat_ids)
print(scat_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.
category_df = pd.DataFrame({'category_id' : cat_ids, 'category': categories})
# Create a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
subcategory_df = pd.DataFrame({'subcategory_id' : scat_ids, 'subcategory': subcategories})
```

```python
# Export categories_df and subcategories_df as CSV files.
category_df.to_csv("Resources/category.csv", index=False)

subcategory_df.to_csv("Resources/subcategory.csv", index=False)
```

Campaign DataFrame
Create a Campaign DataFrame that has the following columns:

The "cf_id" column.
The "contact_id" column.
The company_name column.
The "blurb" column is renamed as "description."
The "goal" column.
The "goal" column is converted to a float datatype.
The "pledged" column is converted to a float datatype.
The "backers_count" column.
The "country" column.
The "currency" column.
The "launched_at" column is renamed as "launch_date" and converted to a datetime format.
The "deadline" column is renamed as "end_date" and converted to a datetime format.
The "category_id" with the unique number matching the category_id from the category DataFrame.
The "subcategory_id" with the unique number matching the subcategory_id from the subcategory DataFrame.
And, create a column that contains the unique four-digit contact ID number from the contact.xlsx file.
Then export the DataFrame as a campaign.csv CSV file.

```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns.
campaign_df = campaign_df.rename(columns={'blurb':'description', 'launched_at':'launch_date', 'deadline':'end_date'})
campaign_df
```

```python
# Convert the goal and pledged columns to a `float` data type.
campaign_df = campaign_df.astype({'goal': 'float','pledged': 'float'})
campaign_df.head()
```

```python
# Check the datatypes
campaign_df.dtypes
```

```python
# Format the launched_date and end_date columns to datetime format
import datetime

campaign_df['launch_date'] = campaign_df['launch_date'].apply(lambda d: datetime.date.fromtimestamp(d))
campaign_df['end_date'] = campaign_df['end_date'].apply(lambda t: datetime.date.fromtimestamp(t))
campaign_df.head()
```

```python
# Merge the campaign_df with the category_df on the "category" column and 
# the subcategory_df on the "subcategory" column.
campaign_merged_df = pd.merge(campaign_df, category_df, how='outer', on=['category'])
campaign_merged_df = pd.merge(campaign_merged_df, subcategory_df, how='outer', on=['subcategory'])
campaign_merged_df.tail(10)
```

```python
# Drop unwanted columns
campaign_cleaned = campaign_merged_df[['cf_id','contact_id','company_name','description','goal','pledged','outcome','backers_count','country','currency','launch_date','end_date','category_id','subcategory_id']]
campaign_cleaned.head()
```

```python
# Export the DataFrame as a CSV file. 
campaign_cleaned.to_csv("Resources/campaign.csv", index=False)
```

Option 1: Use Pandas to create the contacts DataFrame.

```python
import json
dict_values = []

for item in contact_info_df['contact_info']:
    testlist = []
    y = json.loads(item)
    testlist.append(y['contact_id'])
    testlist.append(y['name'])
    testlist.append(y['email'])
    dict_values.append(testlist)

dict_values
```

```python
contact_info_df = pd.DataFrame(dict_values)
contact_info_df
```

```python
contact_info_df = contact_info_df.rename(columns={0:'contact_id', 1:'name', 2:'email'})
contact_info_df
```

```python
contact_info_df.dtypes
```

```python
contact_info_df['first_name'] = ""
contact_info_df['last_name'] = ""
for item in range(len(contact_info_df['name'])):
    splt = contact_info_df['name'][item].split(" ")
    contact_info_df['first_name'][item] = splt[0]
    contact_info_df['last_name'][item] = splt[1]
contact_info_df
```

```python
del contact_info_df['name']
contact_info_df
```

```python
contact_info_df = contact_info_df[['contact_id','first_name','last_name','email']]
contact_info_df
```

```python
# Export the DataFrame as a CSV file. 
contact_info_df.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```

Option 2: Use regex to create the contacts DataFrame.

```python
import re
contact_info_df_copy = pd.read_excel('Resources/contacts.xlsx', header=3)
contact_info_df_copy.head()
```

```python
type(contact_info_df_copy['contact_info'][1])
```

```python
# Extract the four-digit contact ID number.
contact_info_df_copy['contact_id'] = ''
ls = []
for item in contact_info_df_copy['contact_info']:
  ls.append(re.findall('[0-9]{4}', item)[0])
contact_info_df_copy['contact_id'] = ls
contact_info_df_copy.head()
```

```python
# Check the datatypes.
contact_info_df_copy.dtypes
```

```python
# Convert the "contact_id" column to an int64 data type.
contact_info_df_copy['contact_id'] = contact_info_df_copy['contact_id'].astype(int)
contact_info_df_copy.dtypes
```

```python
# Extract the email from the contacts and add the values to a new column.
p = '(?:"name": ")(\s*\w{1,}\s*\w{1,})'
x = pd.DataFrame(contact_info_df_copy['contact_info'].str.extractall(p, flags=re.I))
x.reset_index(inplace = True)
x = x[0]
contact_info_df_copy['name'] = x
contact_info_df_copy.head()
```

```python
# Extract the email from the contacts and add the values to a new column.
q = '(?:"email": ")(.*)(?:"})'
email = pd.DataFrame(contact_info_df_copy['contact_info'].str.extractall(q, flags=re.I))
email.reset_index(inplace = True)
email = email[0]
contact_info_df_copy['email'] = email
contact_info_df_copy.head()
```

```python
contact_info_clean = contact_info_df_copy[['contact_id', 'name', 'email']]
contact_info_clean.head()
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
p = '(\s*\w{1,})(?:\s*\w{1,})'
first_name = pd.DataFrame(contact_info_clean['name'].str.extractall(p, flags=re.I))
first_name.reset_index(inplace = True)
first_name = first_name[0]
contact_info_clean['first_name'] = first_name
contact_info_clean.head()
```

```python
p = '(?:\s*\w{1,}\s)(\s*\w{1,})'
last_name = pd.DataFrame(contact_info_clean['name'].str.extractall(p, flags=re.I))
last_name.reset_index(inplace = True)
last_name = last_name[0]
contact_info_clean['last_name'] = last_name
contact_info_clean.head()
```

```python
# Drop the contact_name column and reorder the columns
contact_info_clean = contact_info_clean[['contact_id','first_name','last_name','email']]
contact_info_clean.head()
```

```python
# Check the datatypes one more time before exporting as CSV file.
contact_info_clean.dtypes
```

```python
# Export the DataFrame as a CSV file. 
contact_info_clean.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```
